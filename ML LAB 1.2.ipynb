```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=45)
imp.fit(data.iloc[:,1:2])
data.iloc[:,1:2]=imp.transform(data.iloc[:,1:2])
```
