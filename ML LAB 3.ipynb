```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dd = sns.load_dataset('titanic')
dd
```

```python
# i. Bar Plot
sns.barplot(x='class', y ='survived', hue ="sex", data=dd)
```

```python
# ii. Count Plot
cp =sns.countplot(x='survived', data=dd)
```

```python
# iii. Facet Grid 
pl =sns.FacetGrid(dd, col="survived")
pl.map(plt.hist,"survived")
```

```python
# iv. Pair Grid
pl =sns.PairGrid(dd)
pl.map(plt.scatter)
```

```python
# v. Relplot
sns.relplot(x='class', y ='survived', data=dd)
```
