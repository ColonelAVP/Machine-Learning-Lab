```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# i. Histogram
df =pd.DataFrame({'length': [1.5,0.5,1.2,0.9,3], 'width': [4.5, 1.5, 1.4, 0.4, 3.6]},index=['A', 'B', 'C', 'D', 'E'])
histo =df.hist(bins =3)
```

```python
# ii. Bar Graphs
df =pd.DataFrame({'x': [1.5,0.5,1.2,0.9,3], 'y': [4.5, 1.5, 1.4, 0.4, 3.6]})
label =['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(df['x'], df['y'], width=1.0, color=['blue', 'red', 'green', 'yellow', 'purple'])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Learning Bar Graph")
plt.show()
```

```python
# iii. Density Plot
df =pd.DataFrame({'length': [1.5,0.5,1.2,0.9,3], 'width': [4.5, 1.5, 1.4, 0.4, 3.6]}, index=['A', 'B', 'C', 'D', 'E'])
df.plot.density(color='green')
plt.title('Learning Density Plot')
plt.show()
```

```python
# v. Box and Whisker Plot
data ={"country": ["Brazil", "Russia", "India", "China", "South Africa"],
       "capital": ["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
       "area": [8.516, 17.10, 3.286, 9.597, 1.221],
       "population": [200.4, 143.5, 1252, 1357, 52.98]}
df =pd.DataFrame(data, columns=["country", "capital", "area", "population"])
plt.boxplot(df[['population', 'area']], showmeans ='True', meanline ='True', vert =True)
```

```python
# vi. Scatter Plot
df =pd.DataFrame([[5.1, 3.5, 0], [4.9, 6.7, 0], [7.0, 3.2, 1], [6.4, 5.3, 7.3], [5.9, 3.0, 2]], 
                 columns=['length', 'width', 'species'])
plot=df.plot.scatter(x='length', y='width', c='Red')
```

```python
# viii. Scatter Matrix plot
np.random.seed(134)                     
N =1000
x1 =np.random.normal(0, 1, N)                       
x2 =x1 +np.random.normal(0, 3, N)              
x3 =2*x1 -x2 +np.random.normal(0, 2, N)
df =pd.DataFrame({'x1':x1,
                  'x2':x2,
                  'x3':x3})
df.head()
```
